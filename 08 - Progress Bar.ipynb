```python
import pandas as pd
from tqdm import tqdm, tqdm_pandas
import numpy as np
from time import sleep
```

```python
def slow_square(x):
    sleep(1)
    return x**2
```

```python
df = pd.DataFrame(np.random.randint(0, 100, (100000, 6)))
tqdm_pandas(tqdm())
```

```python
df.groupby(0).progress_apply(slow_square)
```
