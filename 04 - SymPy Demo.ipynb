```python
import sympy as sym
import numpy as np
from matplotlib.pylab import plt
```

```python
sym.init_printing(use_latex=True)
```

```python
# Register Symbols
sym.var("L K Y A a")
```

```python
y = sym.Symbol('Y')
```

```python
# Cobb-Douglass production function:
Y = ( A*(L**a) ) * ( K ** (1-a) )
```

```python
sym.relational.Eq(y, Y)
```

```python
Ys = Y.subs({A: 10, a:0.6})
```

```python
sym.plotting.plot3d(Ys, (K,0,10), (L,0,10))
```

```python
# Turn sympy symbols into a python function:
Ys_func = sym.lambdify((K, L), Ys, "numpy")
```

```python
# Make 2D permutation list with K=0~10 and L=0~10:
K_n = np.linspace(0,10, 50)
L_n = np.linspace(0,10, 50)
```

```python
result = np.array([[Ys_func(k,l) for l in L_n] for k in K_n])
```

```python
plt.matshow(result)
```
