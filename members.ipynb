```python
import pandas as pd
import seaborn
```

```python
%matplotlib inline
```

```python
members = pd.read_csv('data/PyData-Calgary_Member_List_on_04-02-16.csv', sep='\t')
```

```python
members.sort_values(by=['Joined Group on'], inplace=True)
```

```python
members.groupby(by='Joined Group on')['Member ID'].count().cumsum().plot(figsize=(8,8), fontsize=18)
```
